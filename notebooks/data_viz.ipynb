```python
from pathlib import Path
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_data(ind: int = 0) -> pd.DataFrame:
    """
    Get data from the xy files.
    """
    data_path = Path().resolve().parents[0] / "dataset/input_10K"
    return pd.DataFrame(np.genfromtxt(data_path / f"Bi2Te3-{ind:08d}.xy"
                                      ), columns=['2T', 'I'])
```

```python
idx = pd.read_csv(Path().resolve().parents[0]/ "dataset/input_10K/index_dict.csv")['idx'].values
```

```python
df = get_data(idx[0])
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="2T", y="I", data=df)
plt.show()
```
